# One Hot Encoding using Pandas

```python
# DataFrame containing encoded values of fuel and owner columns

new_df = pd.get_dummies(df,columns=['fuel','owner'], drop_first=True)
new_df
```

```python
new_df.drop(columns=['brand'], inplace=True)
new_df
```

```python
counts = df['brand'].value_counts()
```

```python
threshold = 100

repl = counts[counts <= threshold].index     # Storing the index of brands whose value <= 100

repl
```

```python
df['brand'] = df['brand'].replace(repl, 'uncommon')     # Replacing the brands whose value <= 100 with uncommon

df['brand'].sample(10)
```

```python
# DataFrame containing encoded values of brand column

brand_encoded = pd.get_dummies(df['brand'])
brand_encoded
```

```python
# Merging the DataFrames

df_concat = pd.concat([new_df, brand_encoded], axis=1)
df_concat
```

# One Hot Encoding using Sklearn

```python
X = df.iloc[:,0:4]
y = df.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,random_state=4)

X_train.shape, X_test.shape
```

```python
X_train
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(drop='first',sparse=False,dtype=np.int32)
```

```python
X_train_new = ohe.fit_transform(X_train[['fuel','owner']])
X_test_new = ohe.transform(X_test[['fuel','owner']])
```
